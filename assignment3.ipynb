# Data Cleaning
Remove all emojis

Remove all URLs

Remove all punctuations

```python
import re
import nltk
from nltk.corpus import stopwords  
nltk.download('vader_lexicon')
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))
nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('vader_lexicon')
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
```

```python
# Remove all emojis
def remove_emoji(text):
    text = text.encode('ascii', 'ignore').decode('ascii') #it encodes a unicode string to ascii and ignores errors
    return text
```

```python
# Remove all URLs
def remove_urls(text):
    text = re.sub(r"http\S+", "", text)
    text = re.sub(r"www.(\w+)", "", text)
    return text
```

```python
# Remove all punctuations
def remove_punctuation(text):
    text = re.sub(r'[^\w\s]','',text)
    return text
```

```python
#apply all cleaning functions
dataframe["Descriptions"] = dataframe["Descriptions"] .apply(lambda x: x.lower())
dataframe["Descriptions"] = dataframe["Descriptions"].apply(remove_emoji)
dataframe["Descriptions"] = dataframe["Descriptions"].apply(remove_urls)
dataframe['Descriptions'] = dataframe['Descriptions'].apply(remove_punctuation)
dataframe['Descriptions'] = dataframe['Descriptions'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop_words)]))
```

# 3.[3 pts] Hierarchical clustering implementation:

## a) Implement hierarchical clustering algorithm
i. Generate a distance matrix for to describe the relationship between skills.
ii. Develop a course curriculum based on clustering results (8-12 courses with at least 3 skills/topics covered in each)

Hierarchical clustering is achieved by using Euclidean distance between single observations in data set and linkage criterion. First, calculate the Euclidean distance between each skills. Then using the dendrogram to plot the clustering result with Ward linkage. In order to design 8-12 courses, i set the threshold line at index 13 which gives 8 intersection points with the dendrogram, which indicates 8 clusters.

```python
dataframe_new
```

```python

df_hi = dataframe_new.iloc[:,5:]
df_hi.T.index.tolist()
```

```python

#dist_sin = linkage(df_hi.T, method='ward', metric='euclidean')
plt.figure(figsize=(15,6))
import scipy.cluster.hierarchy as sch
# rotates the x axis labels
# font size for the x axis labels
#Link the clusters
# plot the dendrogram
dendrogram=sch.dendrogram(sch.linkage(df_hi.T, method='ward', metric='euclidean'), labels = df_hi.T.index.tolist(), leaf_rotation=90,color_threshold=13) #Set thereshold at 13)
#keep 8 clusters
plt.axhline(y=13, c='grey', lw=1, linestyle='dashed')
plt.xlabel('skills')
plt.ylabel('Distance')
plt.suptitle("DENDROGRAM SINGLE METHOD",fontsize=18)
plt.show()
     
```


### K-means or DBSCAN clustering implementation
Another method that can help us with unsupervised clustering is k-means clustering. We would assign each observation to the cluster with the closest mean and recompute the means for the observations assigned to each cluster, repeating the steps until they converge. First, to decide how many groups we should use in the algorithm K means. I use the elbow method to find a compromise between inertia and number of groups. We may find that 8 or 9 groups might be appropriate.Then we apply the K-Means algorithm with 9 groups and plot the scatterplot to visualize the results.

```python

from sklearn.cluster import KMeans
data = df_hi
distortions = []



for i in range(1,15):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(data)
    distortions.append(kmeans.inertia_)

plt.plot(range(1,15), distortions, marker='o')
plt.title('Elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
     
```

```python
kmeans = KMeans(n_clusters=8)
kmeans = kmeans.fit(data)
     
```

```python

data['cluster'] = kmeans.fit_predict(data)
data
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
pca_num_components = 2

reduced_data = PCA(n_components=pca_num_components).fit_transform(data)
results = pd.DataFrame(reduced_data,columns=['pca1','pca2'])

sns.scatterplot(x="pca1", y="pca2", hue=data['cluster'], data=results)
plt.title('K-means Clustering with 2 dimensions')
plt.show()
```
